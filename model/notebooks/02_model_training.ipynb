```python
from ultralytics import YOLO
from pathlib import Path
```

## Project Paths

```python
from pathlib import Path

PROJECT_ROOT = Path("../../").resolve()

DEV_DATA_CONFIG = PROJECT_ROOT / "model" / "configs" / "dev_data.yaml"

FULL_DATA_CONFIG = PROJECT_ROOT / "model" / "configs" / "data.yaml"

RUNS_DIR = PROJECT_ROOT / "model" / "runs"

RUNS_DIR.mkdir(exist_ok=True)
```

## Loading the Pretrained Model

```python
model = YOLO("yolo11n.pt")
```

## Training Configuration

```python
training_config = {
    "data": str(DEV_DATA_CONFIG),
    "epochs": 5,
    "imgsz": 640,
    "batch": 8,
    "device": "cpu",
    "workers": 4,
    "project": str(RUNS_DIR),
    "name": "baseline_yolo11n",
    "pretrained": True,
    "verbose": True
}

training_config
```

## Model Information

```python
model.info()
```

## Baseline Training

The following section trains the baseline YOLO11 Nano model using the RDD2022 dataset.

This initial experiment focuses on validating the complete training pipeline rather than maximizing model accuracy.

## Training Strategy

The baseline model is trained to validate the complete RoadVision training pipeline rather than maximize detection performance.

### Model Selection

The **YOLO11 Nano (YOLO11n)** model is used as the baseline because it provides a good balance between training speed and detection performance during development. A larger model (YOLO11s) will be used later for the final production training.

### Image Resolution

An input image size of **640 × 640 pixels** is selected for the baseline experiment. During dataset exploration, most sampled images were found to have resolutions between **512 px and 720 px**, making 640 × 640 an appropriate compromise between computational efficiency and spatial detail.

### Batch Size

A batch size of **8** is used to ensure stable CPU memory usage on the local development machine (16 GB RAM). Larger batch sizes will be used during GPU-based production training.

### Epochs

The baseline model is trained for **5 epochs** only. The objective of this experiment is to verify the complete training pipeline, including data loading, model optimization, checkpoint generation, and evaluation. The final production model will be trained for significantly more epochs on the complete RDD2022 dataset using a GPU.

### Data Augmentation

Default YOLO11 augmentation settings are used during the baseline experiment. Additional augmentation techniques and hyperparameter tuning will be explored after establishing a reliable baseline model.

```python
results = model.train(
    **training_config
)
```

## Training Outcome

The baseline YOLO11 Nano model was successfully trained on the RoadVision development dataset.

The training process generated model checkpoints and evaluation artifacts, including training metrics, model weights, and performance visualizations. These outputs will be analyzed in detail in the subsequent model evaluation notebook.

```python
import pandas as pd

results_csv = RUNS_DIR / "baseline_yolo11n" / "results.csv"

training_results = pd.read_csv(results_csv)

summary = pd.DataFrame({
    "Metric": [
        "Epoch",
        "Training Box Loss",
        "Training Classification Loss",
        "Training DFL Loss",
        "Precision",
        "Recall",
        "mAP@50",
        "mAP@50-95"
    ],
    "Value": [
        training_results.iloc[-1]["epoch"],
        training_results.iloc[-1]["train/box_loss"],
        training_results.iloc[-1]["train/cls_loss"],
        training_results.iloc[-1]["train/dfl_loss"],
        training_results.iloc[-1]["metrics/precision(B)"],
        training_results.iloc[-1]["metrics/recall(B)"],
        training_results.iloc[-1]["metrics/mAP50(B)"],
        training_results.iloc[-1]["metrics/mAP50-95(B)"]
    ]
})

summary
```
